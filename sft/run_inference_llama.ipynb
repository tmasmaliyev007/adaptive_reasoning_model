# **Load Libraries**

```python
from llama_cpp import Llama
from llama_cpp.llama_chat_format import Jinja2ChatFormatter

from inference.llama import run_inference
```

# **Initialize Model**

```python
# Configure model settings
model_name = "../model/gguf/qwen-3-4B/b16-full.gguf"

chat_template_name = "../template/qwen3.jinja"
MAX_SEQ_LENGTH = 20000
```

```python
# Load the model
llm = Llama(
    model_path = model_name,
    n_gpu_layers = 0,
    verbose = False,
    flash_attn = True,
    n_ctx = MAX_SEQ_LENGTH
)
```

```python
# Load Template
with open(chat_template_name, 'r') as f:
    chat_template = f.read()

# Load formatter
formatter = Jinja2ChatFormatter(
    template=chat_template,
    add_generation_prompt=True,
    bos_token=llm._model.token_get_text(llm.token_bos()),
    eos_token=llm._model.token_get_text(llm.token_eos())
)
```

# **Run Inference**

```python
# ── Single instruction example ────────────────────────────────────────────────
instruction = "Solve the following equation and explain each step: 3x + 7 = 22"

response = run_inference(instruction, formatter, max_new_tokens=1024, temperature=0.7)
print("=" * 60)
print("INSTRUCTION:")
print(instruction)
print("=" * 60)
print("RESPONSE:")
print(response)
```
